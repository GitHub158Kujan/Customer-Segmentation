# EDA

```python
fig,axs=plt.subplots(1,3,figsize=(20,5))
sns.histplot(data=df,x='Age',kde='True',color='blue',ax=axs[0]).set(title='Age Distribution')
sns.histplot(data=df,x='Annual Income (k$)',kde=True,color='red',ax=axs[1]).set(title='Annual Income Distribution(ks)')
sns.histplot(data=df,x='Spending Score (1-100)',kde=True,color='green',ax=axs[2]).set(title='Spending Score(1-100) Distribution')
plt.show()
```

```python
df_ = df.groupby('Gender').size()
df_.plot(kind='pie',subplots=True,colors=['green','blue'],explode=[0,0.001],labels=['Female','Male'],autopct='%0.1f%%')
plt.title("Gender Distribution")
plt.show()
```

```python
sns.kdeplot(data=df, x='Age', hue='Gender', fill=True).set(title='Age Distribution by Gender')
plt.show()
```

```python
bins = [0, 20, 30, 40, 50, 100]  
labels = ['<20', '20-30', '30-40', '40-50', '50+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
sns.barplot(data=df,x='Age Group',y='Spending Score (1-100)',hue='Gender',errorbar=None).set(title='Spending Score by Age')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df,x='Annual Income (k$)',y='Spending Score (1-100)',hue='Gender',s=200)
plt.title("Relationship Between Annual Income and Spending Score")
plt.show()
```

```python
plt.figure(figsize=(10,7))
for gender in ['Male','Female']:
  plt.scatter(x='Age',y='Annual Income (k$)',data=df[df['Gender']==gender],
             s=200,alpha=0.5, label=gender)
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('Age VS Annual Income')
plt.legend()
plt.show()
```

# K-means clustering

```python
X=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans
```

```python
k_means=KMeans(n_clusters=4)
k_means.fit_predict(X)
```

**Elbow Method**

```python
wcss=[]
for i in range(1,11):
   k_means=KMeans(n_clusters=i)
   k_means.fit_predict(X)
   wcss.append(k_means.inertia_)

    
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.title('Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
```

**Model Training**

```python
X=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
k_means=KMeans(n_clusters=5,random_state=42)
y_means=k_means.fit_predict(X)
```

```python
y_means
```

```python
plt.scatter(X.iloc[y_means==0,0],X.iloc[y_means==0,1],color='red',label='Cluster 1')
plt.scatter(X.iloc[y_means==1,0],X.iloc[y_means==1,1],color='blue',label='Cluster 2')
plt.scatter(X.iloc[y_means==2,0],X.iloc[y_means==2,1],color='green',label='Cluster 3')
plt.scatter(X.iloc[y_means==3,0],X.iloc[y_means==3,1],color='black',label='Cluster 4')
plt.scatter(X.iloc[y_means==4,0],X.iloc[y_means==4,1],color='magenta',label='Cluster 5')
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1],color='yellow')
plt.title('Customer Segmentation')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
```

<b>Cluster Analysis:</b><br>
The plot shows five distinct customer clusters based on their income levels and spending patterns:<br>
<b>Cluster 1 (Red):</b> Customers with moderate income and average spending. This group represents balanced spenders.<br>
<b>Cluster 2 (Blue):</b> High-income customers with high spending. These are likely the premium customers.<br>
<b>Cluster 3 (Green):</b> Low-income customers with high spending. They might represent budget-conscious but frequent shoppers.<br>
<b>Cluster 4 (Black):</b> High-income customers with low spending. These may be selective spenders or individuals who save more.<br>
<b>Cluster 5 (Pink):</b> Low-income customers with low spending. These customers have minimal engagement or financial constraints.<br>

**3-D Plot**

```python
import plotly.express as px
```

```python
X=df[['Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
wcss=[]
for i in range(1,11):
   k_means=KMeans(n_clusters=i)
   k_means.fit_predict(X)
   wcss.append(k_means.inertia_)
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.title('Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
```

```python
k_means=KMeans(n_clusters=5)
y_predict=k_means.fit_predict(X)
```

```python
labels = k_means.labels_
df['Cluster'] = labels
fig = px.scatter_3d(
    df,
    x='Age',
    y='Annual Income (k$)',
    z='Spending Score (1-100)', 
    color='Cluster',
    title="3D Scatter Plot of K-Means Clusters",
    labels={
        'Age': 'Age',
        'Annual Income (k$)': 'Annual Income',
        'Spending Score (1-100)': 'Spending Score'
    }
)
fig.update_layout(
    width=800,  
    height=600  
)

fig.show()
```

<b>Cluster Insights:</b><br>
The <b>yellow cluster</b>Individuals in this group tend to have high annual income but low spending scores. They appear to span across a wide range of ages, suggesting that age does not heavily influence spending behavior for this group.<br>
The <b>orange cluster</b> represents individuals with high annual income and high spending scores. This may indicate a trend of higher spending propensity among younger, high-income earners.<br>
The <b>purple cluster</b> corresponds to low annual income and low spending scores.They are predominantly older, indicating that this segment might include more conservative spenders or retirees.<br>
The <b>blue cluster</b> includes individuals with varying spending scores but moderate income levels.This cluster shows a mix of ages from 18 to 40.<br>
The <b>pink cluster</b> might consist of individuals with low-to-moderate income and varying spending scores.This cluster shows a balanced mix of ages, suggesting diverse spending habits across age groups.<br>
